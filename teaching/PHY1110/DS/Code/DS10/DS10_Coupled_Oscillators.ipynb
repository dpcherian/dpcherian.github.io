```python
# Importing the necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Dynamics ####################

# Constants: ##############

k1 = 10
k2 = 1
m1 = 1
m2 = 1


# Acceleration function: ##
def a(eqn, pos_x1, pos_x2):
    # If we are looking for the acceleration of mass 1, eqn == 1, otherwise, eqn == 2
    if(eqn==1):
        acc = -(k1/m1)*pos_x1 + (k2/m1)*(pos_x2 - pos_x1)
    elif(eqn==2):
        acc = -(k1/m2)*pos_x2 + (k2/m2)*(pos_x1 - pos_x2)
    
    return acc

#################################
```

```python
# Kinematics ####################

# Initial Conditions : #####

x1_0 = 1
x2_0 = 0

v1_0 = 0
v2_0 = 0


# Arrays : #################

dt = 0.01
T  = 20
N = int(T/dt)

t  = np.zeros(N+1,float)

x1 = np.zeros(N+1,float)
x2 = np.zeros(N+1,float)
v1 = np.zeros(N+1,float)
v2 = np.zeros(N+1,float)


# Leapfrog : ###############

x1[0] = x1_0
x2[0] = x2_0

v1[0] = v1_0 + a(1,x1_0,x2_0)*dt/2
v2[0] = v2_0 + a(2,x1_0,x2_0)*dt/2

# Kinematics loop : ########

for i in range(1,N+1):
    t[i]  = i*dt
    
    x1[i] = x1[i-1] + v1[i-1]*dt
    x2[i] = x2[i-1] + v2[i-1]*dt
    
    v1[i] = v1[i-1] + a(1,x1[i],x2[i])*dt
    v2[i] = v2[i-1] + a(2,x1[i],x2[i])*dt

############################################
```

```python
# Plotting #################################

plt.plot(t,x1,label='$x_1$', color='b')
plt.xlabel('$t$')
plt.ylabel('$x$')
plt.legend()
plt.show()

plt.plot(t,x2, label = '$x_2$',color='r')
plt.xlabel('$t$')
plt.ylabel('$x$')
plt.legend()
plt.show()

plt.plot(t,x1+x2, label = '$x_1 + x_2$')
plt.plot(t,x1-x2, label = '$x_1 - x_2$')
plt.xlabel('$t$')
plt.ylabel('$x$')
plt.legend()
plt.show()

############################################
```
