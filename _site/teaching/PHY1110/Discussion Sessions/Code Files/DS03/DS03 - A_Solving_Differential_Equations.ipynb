```python
# Importing the necessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the dynamics of the system. The acceleration of the system `a`
# accepts two variables (in this case, the position `x` and velocity `v`) 
# and returns `-x`. It is important to note that in this process, the variable
# types of `x` and `v` are unimportant. They could be numbers, numpy arrays, 
# VPython vectors and so on.
def a(x,v):
    return -x   # While the velocity does not affect the result in this program, 
                # we have left it in to account for cases which might. 
```

```python
t = 0                       # Initial time
tf= 10                      # Final time
dt = 0.01                   # Time step (this should be chosen carefully)

N = int((tf-t)/dt)          # The total number of steps in the time interval

x = np.zeros(N+1,float)     # A numpy array of zeros of length N+1 for position
v = np.zeros(N+1,float)     # A numpy array of zeros of length N+1 for velocity
t = np.zeros(N+1,float)     # A numpy array of zeros of length N+1 for time


# Setting up the initial conditions
#-----------------------------------#
x0 = 1
v0 = 0

x[0] = x0                    # Assigning these initial conditions to the arrays
v[0] = v0 + a(x0,v0)*dt/2    # Setting the initial velocity using the leapfrog method
#-----------------------------------#


for i in range(1,N+1):       # The loop that determines the kinematics by incrementing
    t[i] = i*dt              # each of the above arrays by their corresponding increases.
    x[i] = x[i-1] + v[i-1]*dt
    v[i] = v[i-1] + a(x[i],v[i-1])*dt
```

```python
plt.scatter(t,x)             # Plot a scatter plot of x vs. t
xexact = 1*np.cos(t)         # An array of the exact solution to a simple harmonic oscillator
plt.plot(t,xexact,color='r') # Plot the exact solution as a function of time
plt.show()                   # Show both these plots above on the same figure. 
```
